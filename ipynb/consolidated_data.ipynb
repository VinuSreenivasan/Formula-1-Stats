```python
import pandas as pd
import numpy as np
```

```python
circuits_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/circuits.csv')
constructors_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/constructors.csv')
constructors_results_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/constructorResults.csv')
cons_std_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/constructorStandings.csv')
drivers_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/drivers.csv')
driver_standings_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/driverStandings.csv')
races_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/races.csv')
results_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/results.csv')
status_data = pd.read_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/status.csv')
```

```python
races_result = pd.merge(races_data[['raceId','year','round','circuitId','name','date']], results_data[['raceId','driverId','constructorId','number','grid','position','positionText','laps','points', 'statusId']], on='raceId')
races_result
```

```python
races_results_circuit = pd.merge(races_result,circuits_data[['circuitId','circuitRef','name','location','country', 'lat','lng']], on="circuitId")
races_results_circuit
```

```python
races_results_circuit_drivers = pd.merge(races_results_circuit, drivers_data[['driverId','driverRef','number','forename','surname','dob','nationality']], on='driverId')
races_results_circuit_drivers
```

```python
races_results_circuit_drivers.rename(columns={'year':'season',
                          'name_x':'raceName',
                          'circuitRef':'circuitId_x','name_y':'circuitName'}, 
                 inplace=True)
```

```python
races_results_circuit_drivers
```

```python
races_results_circuit_drivers.rename(columns={'dob':'dateOfBirth','forename':'givenName','surname':'familyName','grid':'results.grid','laps':'results.laps'}, 
                 inplace=True)
```

```python
races_results_circuit_drivers
```

```python
races_results_circuit_drivers_cons = pd.merge(races_results_circuit_drivers, constructors_data[['constructorId','constructorRef','name','nationality']], on="constructorId")
races_results_circuit_drivers_cons
```

```python
races_results_circuit_drivers_cons.rename(columns={'nationality_x':'nationality','nationality_y':'nationality.1','name':'name_x'}, 
                 inplace=True)
races_results_circuit_drivers_cons
```

```python
races_results_circuit_drivers_cons['statusId']
```

```python
races_results_circuit_drivers_cons_status = pd.merge(races_results_circuit_drivers_cons, status_data, on='statusId')
races_results_circuit_drivers_cons_status
```

```python
races_results_circuit_drivers_cons_status.rename(columns={'status':'results.status'}, inplace=True)
races_results_circuit_drivers_cons_status
```

```python
races_results_circuit_drivers_cons_status.columns
```

```python
#races_results_circuit_drivers_cons_status=races_results_circuit_drivers_cons_status.drop(['statusId'], axis=1)
races_results_circuit_drivers_cons_status
```

```python
races_results_circuit_drivers_cons_status = races_results_circuit_drivers_cons_status.drop(['number_y','circuitId','constructorId', 'driverId'], axis=1)
races_results_circuit_drivers_cons_status
```

```python
races_results_circuit_drivers_cons_status.columns
```

```python
races_results_circuit_drivers_cons_status.rename(columns={'number_x':'number'}, inplace=True)
```

```python
races_results_circuit_drivers_cons_status = races_results_circuit_drivers_cons_status[['season','round','raceName','date','circuitId_x','circuitName','number','position','positionText','points','driverRef','givenName','familyName','dateOfBirth','nationality','constructorRef','name_x','nationality.1','results.grid','results.laps','results.status','location','country','lat','lng']]
races_results_circuit_drivers_cons_status.columns
```

```python
races_results_circuit_drivers_cons_status = races_results_circuit_drivers_cons_status.sort_values(['season','round'], ascending=[True, True])
races_results_circuit_drivers_cons_status
```

```python
races_results_circuit_drivers_cons_status.to_csv('C:/Users/sanja/Desktop/Courses/Data vis/Project/formula-1-race-data-19502017/consolidated_f1_stats.csv')
```
